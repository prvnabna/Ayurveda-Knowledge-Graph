```python
from langdetect import detect
import pandas as pd

with open("cleaned_text_v5.txt", "r", encoding="utf-8") as f:
    text = f.read()

detected = detect(text[:5000])

df = pd.DataFrame({
    "Document": ["AyUG-SN AI-2"],
    "Language": [detected]
})

df.to_csv("language_labels.csv", index=False)

print(df)
```
