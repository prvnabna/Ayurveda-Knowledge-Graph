```python
import fitz

pdf = fitz.open("/content/AyUG-SN AI-2.pdf")

text = ""

for page in pdf:
    text += page.get_text()

with open("raw_text.txt", "w", encoding="utf-8") as f:
    f.write(text)

print("raw_text.txt created successfully")
```

```python
import re

with open("raw_text.txt", "r", encoding="utf-8") as f:
    text = f.read()

# Remove multiple spaces
text = re.sub(r'\s+', ' ', text)

# Remove repeated newlines
text = re.sub(r'\n+', '\n', text)

# Remove leading/trailing spaces
text = text.strip()

with open("cleaned_text.txt", "w", encoding="utf-8") as f:
    f.write(text)

print("cleaned_text.txt created")
```

```python
import re

with open("raw_text.txt", "r", encoding="utf-8") as f:
    text = f.read()

# Fix PDF ligatures
text = text.replace("ﬁ", "fi")
text = text.replace("ﬂ", "fl")

# Remove URLs
text = re.sub(r'http\S+', '', text)

# Remove extra spaces
text = re.sub(r'\s+', ' ', text)

# Remove leading/trailing spaces
text = text.strip()

with open("cleaned_text_v3.txt", "w", encoding="utf-8") as f:
    f.write(text)

print("cleaned_text_v3.txt created successfully")
```

```python
with open("cleaned_text_v3.txt", "r", encoding="utf-8") as f:
    text = f.read()

print(text[:3000])
```

```python
with open("cleaned_text_v3.txt", "r", encoding="utf-8") as f:
    text = f.read()

# Find where abstract starts
start = text.find("Abstract:")

if start != -1:
    text = text[start:]

with open("cleaned_text_v4.txt", "w", encoding="utf-8") as f:
    f.write(text)

print("cleaned_text_v4.txt created")
```

```python
with open("cleaned_text_v4.txt", "r", encoding="utf-8") as f:
    text = f.read()

print(text[:1500])
```

```python
import re

with open("cleaned_text_v4.txt", "r", encoding="utf-8") as f:
    text = f.read()

# Remove page numbers like "2 of 34"
text = re.sub(r'\b\d+\s+of\s+\d+\b', '', text)

# Remove journal headers
text = re.sub(r'Pharmaceutics\s+\d{4},\s*\d+,\s*\d+', '', text)

# Remove figure captions references
text = re.sub(r'Figure\s+\d+\.*', '', text)

# Remove extra spaces again
text = re.sub(r'\s+', ' ', text)

text = text.strip()

with open("cleaned_text_v5.txt", "w", encoding="utf-8") as f:
    f.write(text)

print("cleaned_text_v5.txt created")
```
